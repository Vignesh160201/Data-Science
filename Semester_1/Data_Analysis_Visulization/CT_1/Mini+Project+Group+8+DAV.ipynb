**Convert the column Level from categorical type to numerical type (1 mark)**

```python
#convert the column Level from categorical type to numerical type
df=df.replace({'Level':{'Low': 1, 'Medium': 2, 'High': 3}})
```

**Import the vizualization library (1 mark)**

```python
#import the vizualization library
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
```

**Display cor-relation matrix (1 mark)**

```python
#display cor-relation matrix
plt.figure(figsize = (18,9))
sns.heatmap(df.corr(), cmap='GnBu', annot=True)
plt.show()
```

**Display the histogram of the dataframe (2 mark)**

```python
#display the histogram of the dataframe
df.hist(figsize=(15, 15))
```

Display the countplot for Gender at different Level (1 mark)

```python
#display the countplot for Gender at different Level
plt.title('Gender Count')
sns.countplot(data=df, x = 'Gender', hue = 'Level', palette = 'Set1')
```

**Display the countplot for Gender at different Smoking (1 mark)**

```python
#display the countplot for Gender at different Smoking
plt.title('Smoking Count')
sns.countplot(data=df, x = 'Gender', hue = 'Smoking', palette = 'Set1')
```

**Display a pie chart the % of different Level (1 mark)**

```python
#display a pie chart the % of different Level
plt.figure(figsize = (11, 9))
plt.title("Lung Cancer Chances Due to Air Polution")
plt.pie(df['Level'].value_counts(), explode = (0.1, 0.02, 0.02), labels = ['High', 'Medium', 'Low'], autopct = "%1.2f%%", shadow = True)
plt.legend(title = "Lung Cancer Chances", loc = "lower left")
```

**Display a distribution plot for Level column (1 mark)**

```python
#display a distribution plot for Level column
sns.displot(df['Level'], kde=True)
```

**Display a line plot between Age and Level (1 mark)**

```python
#display a line plot between Age and Level
sns.lineplot(data=df,x='Age',y='Level')
```

**Display a line plot between Chest Pain and Level (1 mark)**

```python
#display a line plot between Chest Pain and Level
sns.lineplot(data=df,x='Chest Pain',y='Level')
```

**Display a scatter plot between Smoking and Shortness of Breath w.r.t Level (hue) (1 mark)**

```python
#display a scatter plot between Smoking and Shortness of Breath w.r.t Level (hue)
sns.scatterplot(data=df,x='Smoking',y='Shortness of Breath',hue='Level')
```

**Display a scatter plot between Age and Gender w.r.t Level (hue) (1 mark)**

```python
#display a scatter plot between Age and Gender w.r.t Level (hue)
sns.scatterplot(data=df,x='Age',y='Gender',hue='Level')
```

**display a histogram plot between OccuPational Hazards and Gender (1 mark)**

```python
#display a histogram plot between OccuPational Hazards and Gender
plt.subplots(figsize=(20, 7))

sns.histplot(data=df, x=df["OccuPational Hazards"], hue=df["Gender"], multiple="dodge", shrink=.8,palette=["pink","darkblue"]).set(title='Relation Between OccuPational Hazards and Gender')
```

**Display a bar plot for obesity and coughing of blood (1 mark)**

```python
#display a bar plot for obesity and coughing of blood
plt.figure(figsize=(18, 6))
plt.title("Relation between Obesity and Coughing of Blood")
sns.barplot(data=df, x="Obesity", y="Coughing of Blood",palette="husl")
```
